# TCP Congestion Control Simulation
## Comparison of TCP Tahoe vs TCP Reno

This notebook demonstrates the behavior of TCP Tahoe and TCP Reno congestion control algorithms using a simple simulation.

It shows how congestion window (cwnd) changes over time, especially during packet loss.

### Key Concepts

- **TCP Tahoe**:
  - Slow Start
  - Congestion Avoidance
  - On packet loss → cwnd resets to 1

- **TCP Reno**:
  - Similar to Tahoe
  - On packet loss → cwnd reduces to half (not reset)

```python
import matplotlib.pyplot as plt

time = list(range(1, 21))

# Tahoe
tahoe = []
cwnd = 1
for t in time:
    tahoe.append(cwnd)
    if cwnd < 16:
        cwnd *= 2
    else:
        cwnd += 1
    if t == 10:
        cwnd = 1

# Reno
reno = []
cwnd = 1
for t in time:
    reno.append(cwnd)
    if cwnd < 16:
        cwnd *= 2
    else:
        cwnd += 1
    if t == 10:
        cwnd = cwnd // 2

plt.plot(time, tahoe, marker='o', label="TCP Tahoe")
plt.plot(time, reno, marker='s', label="TCP Reno")

plt.xlabel("Time (RTT)")
plt.ylabel("Congestion Window (cwnd)")
plt.title("TCP Tahoe vs Reno")
plt.legend()
plt.grid()

plt.show()
```

### Conclusion

The simulation shows that TCP Reno performs better than Tahoe after packet loss because it avoids resetting the congestion window completely.

This leads to better bandwidth utilization and faster recovery.
